## Web Scraping HDFC Credit Card Website

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.hdfcbank.com/personal/pay/cards/credit-cards'
response = requests.get(url)
data = BeautifulSoup(response.text, 'html.parser')
print(data.prettify())
```

## name of each card

```python
name=[]
card = data.find_all(class_='card-name')
for i in card:
    name.append(i.text.strip())
name
```

```python
c_know = data.find_all(class_='btnParent')
for i in c_know:
    d1 = i.find(class_='btn btn-primary btn-custom btn-primary-customs hide-for-small mybtn normal-url')
    print(d1)
```

## tried to get the url for know More button but actually getiing url for both know more and Apply online button

```python
c_know = data.find_all(class_='btn btn-primary btn-custom btn-primary-customs hide-for-small mybtn normal-url')
for i in c_know:
    print(i['href'])
```

## getting the details of each card 

```python
c= data.find_all(class_='col-sm-8 parentdiv')
info=[]
for i in c:
    info.append(list(i.stripped_strings))
info
```

```python
d= data.find_all(class_='col-sm-8 parentdiv')
url=[]
for i in d:
    f= i.find_all('a',class_="btn btn-primary btn-custom btn-primary-customs show-for-small hide-for-large normal-url")
    for j in f:
        url.append(j['href'])
```

```python
c= data.find_all(class_="offer-dtl mB10")
info=[]
for i in c:
    info.append(list(i.stripped_strings))
info
```
